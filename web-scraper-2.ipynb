```python
singles_df = singles_df.copy()
def extract_oz(row):
    if 'oz' in row['name']:
        match = re.search(r'(\d+(?:\.?\d+)?)\s*oz', row['name'], re.IGNORECASE)
        if match:
            number = match.group(1)
            if '.' in number:
                return float(number)
            else:
                return int(number)
    return None

singles_df['oz'] = singles_df.apply(extract_oz, axis=1)
with pd.option_context('display.max_rows', None, 'display.max_colwidth', None):

        singles_df[singles_df['oz'].isna()]
```
